## Part 1

We want to check the integrity of an orbits map.

The map describes objects orbiting around others.

> B orbits around A: A)B

> B and C orbits around A :
> ```
> A)B
> A)C
> ```

> B and C orbits around A, D orbits around C:
> ```
> A)B
> A)C
> C)D
> ```

To check the integrity of a map, we must sum the total count of direct and indirect orbits.
In the previous example, there is:
* 3 direct obrits: A)B, A)C and C)D
* 1 indirect orbit: A)D
* the checksum is 4

All map start with a central point, COM, not orbiting around anything else.

### Reading a map

````python
from collections import defaultdict

def read_map(map_txt):
    """Read a map in text format and interpret it.
    
    Return a list of list.
    """
    # x's children are all objects orbiting around x.
    # We will store all objects and their children.
    objects = defaultdict(dict)
    
    # COM is the source of everything
    result_map = {"COM": objects["COM"]}

    for line in map_txt.splitlines():
        if not ")" in line:
            continue

        # b orbits around a
        a, b = line.split(")")
        
        a_children = objects[a]
        a_children[b] = objects[b]
        
    return result_map
````

### Checksum

````python
def orbits(map_):
    """Return the number of orbits in a map.
    
    Return a tuple (x, y, z):
    * x : total of direct orbits
    * y : total of indirect orbits
    * z : total of objects not orbiting
    """
    direct = 0
    indirect = 0
    total = 0

    for obj in map_:
        sub_direct, sub_indirect, sub_total = orbits(map_[obj])

        direct += sub_direct + sub_total
        indirect += sub_direct + sub_indirect
        total += 1        

    return direct, indirect, total
````

## Part 2

We known need to compute the length of the shortest path between us (object YOU) and Santa (object SAN).

````python
def get_path(map_, dest):
    """Return a path from the origin to the object `dest`.
    
    The path is a list of objects. If `dest` is not in `map_`, return an empty list.
    
    The path is actually returned reversed: from dest to the source.
    """
    for obj in map_:
        if obj == dest:
            return [dest]
        
        path = get_path(map_[obj], dest)
        if path:
            path.append(obj)
            return path
        
    return []
````

Let test it works as expected:

````python
map_ = {
    "A": {
        "B": {},
        "C": {
            "D": {},
            "E": {
                "F": {},
                "G": {},
            },
            "H": {},
        },
        "I": {
            "J": {},
        },
        "K": {},
    },
}

print(get_path(map_, "A"))
print(get_path(map_, "G"))
print(get_path(map_, "J"))
print(get_path(map_, "COM"))
````

Now we need to compute the shortest path between two objects.
We will first compute the path from the source to each object, then we will find closest common ancestor and use it to compute the final path.

````python
def get_path_between(map_, src, dest):
    """Return the shortest path between `src` and `dst`."""
    path_to_src = get_path(map_, src)
    path_to_dest = list(reversed(get_path(map_, dest)))
    
    for i, obj in enumerate(path_to_src):
        if obj in path_to_dest:
            break
    else:
        # no common ancestor found
        return []
    
    return (path_to_src[:i]) + path_to_dest[path_to_dest.index(obj):]
````
